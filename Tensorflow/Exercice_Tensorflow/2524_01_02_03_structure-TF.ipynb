## Les tenseurs  de TensorFlow 
### Tenseurs
TensorFlow tire son nom des **tenseurs** (**tensors** en anglais).

Ce sont des tableaux de dimension arbitraire. 
Un tenseur est la généralisation d'une matrice.
L'intérêt des tenseurs devient clair lorsque l'on manipule des données complexes dans ses programmes de Machine Learning (images, vidéo, etc).


### Constantes et variables
Les tenseurs sont stockés sous forme de **constantes** ou de **variables**. 

Les **constantes** contiennent des tenseurs dont les valeurs ne peuvent *pas* changer.
Les **variables** contiennent des tenseurs dont les valeurs peuvent changer. 


Testons sans plus attendre, *constantes* et *variables*



### Graphe
Un graphe TensorFlow appelé également graphe de calcul (*graph* en anglais) est  une structure de données. C'est une façon de représenter les calculs. 
Un graphe est constitué de noeud et d'arètes.

les  **nœuds** d'un graphe de calculs sont des opérations.
Dans TensorFlow, chaque opération est associée à un graphe. 
**Les arcs** d'un graph sont des tenseurs. 


---


On peut voir le déroulement d'un programme Tensorflow comme un flot dans le **graphe**.
Ceci explique le choix du nom *Tensor-flow*.

Les tenseurs s'écoulent tout du long du graphe et subissent à chaque noeud une **opération**.

On  enchaine les opérations de la manière suivante :
Le tenseur de sortie d'une opération est le tenseur d'entrée de l'opération suivante. 


---


Les tenseurs sont stockés dans le graphe de calculs sous forme de **constantes** ou de **variables**. 



### Execution d'un graphe de calculs
TensorFlow implémente un modèle d'exécution parcimonieuse  (**"lazy execution model"**), ce qui signifie que les nœuds ne sont calculés qu'au besoin, c'est-à-dire, selon les besoins des nœuds associés

Pour provoquer l'exécution des calculs, nous réalisons les opérations suivantes : 
* ouverture d'une **session**
* initialisation
* lancement de la session

## Compléments sur les Graphes de calcul
Dans tout programme TensorFlow, il y a 2 phases distinctes. La phase de construction du graph et ensuite la phase d'exécution du graph.

Commençons par la phase de construction
### Phase de construction du graph de calculs
Les variables et les constantes sont déclarées dans cette phase.

Le **plus important** est de  comprendre que le code de déclaration des variables et des constantes ne génère **aucun** calcul ! 

Deux exemples : 

Le code suivant ne génère aucun calcul :
```
import tensorflow as tf
a = tf.constant(3, name="a") 
b = tf.Variable(4, name="b") 

```

Le code suivant ne génère aucun calcul :
```
f = a*a*b + b + 2
```



La seule chose qui est faite par ce code est de **créer** l'arbre de calcul, le graph.
Une image pour comprendre : on commence par dessiner, créer le réseau avant d'y faire transiter  les calculs


La phase de construction du graph pose l'architecture mais ne réalise aucun calculs.

---

**Ceci est le point le plus difficile à retenir lorsque l'on démarre l'apprentissage de TensorFlow**

---




### Phase d'exécution d'un graph de calculs
La phase d'exécution, en règle générale, est constituée par l'optimisation du modèle de Machine Learning. 

Par exemple, les poids des connections pour un algorithme de Deep Learning . 

Concrètement, le plus souvent, cela prend la forme d'une boucle. On optimise les paramètres à chaque itération de la boucle. 

Voyons à présent le code nécessaire à **l'évaluation** de l'arbre de calcul

* initialiser les variables  : ```
tf.global_variables_initializer()
```


*  ouvrir une **session**  : ```
with tf.Session() as sess:
```
* exécuter l'initialisation : ```
init.run()
```
* et enfin évaluer les variables si nécessaire : ```
eval()
```

Un exemple :




````
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

a = tf.constant(3, name="a") 
b = tf.Variable(4, name="b")

f = a*a*b + b + 2

# prépare initialisation
init = tf.global_variables_initializer()

# ouverture session
with tf.Session() as sess:
  # lancer init 
  init.run() 

  result = f.eval()
print (result)
````

Durant la phase d'exécution, TensorFlow s'occupe de placer les **opérations** et de "déporter" le calcul sur le matériel adéquat (CPU ou carte graphique GPU) et enfin de réaliser ces calculs.


Remarque :

le code suivant
```
tf.global_variables_initializer()
```
crée un **noeud** dans le graph qui permettra d'initialiser toutes les variables lorsque l'arbre de calcul sera calculé.

Ce sont des tableaux de dimension arbitraire. 
Un tenseur est la généralisation d'une matrice.
L'intérêt des tenseurs devient clair lorsque l'on manipule des images.


Voici quelques exemples de tenseurs :

* Un scalaire, un chiffre, est un tableau 0-d (un tenseur d'ordre 0). 

Par exemple, 3
* Un vecteur est un tableau 1D (un tenseur de 1er ordre). 

Par exemple,`[3 , 14 , 15]` .
* Une matrice est un tableau 2D  (un tenseur d'ordre 2). 

Par exemple,` [[9, 5 , -4][4 , 7 , 0]].`

il peut y avoir des tenseurs de dimensions plus élevé. Par exemple, 
* image N&B sera encodée par un tenseur 3D  : ```
 [hauteur en pxl, largeur en pxl , channel]
 ```
 
Pour une image N&B, *channel* = 1. Pour une image couleur, *channel*= 3 -> (R,G,B) 

Exemple : une image N&B de dimension (en pixel) 28  x 28  : ```
[28 , 28 , 1] ```

Exemple : une image couleur de dimension (en pixel) 28  x 28 : ```
[28 , 28 , 3] ```

Exemple : 60,000 images N&B de dimension (en pixel) 28  x 28  : ```
[60000 , 28 , 28 , 1] ```

### Alimenter les modèles en data  avec Placeholder

Voyons ensemble comment TensorFlow "alimente" les modèles de calcul avec les données !

Dans l'immense majorité des cas, les algorithmes de Machine Learning sont alimentés par "paquets" de données (le terme technique utilisé est : *mini batch*).

A chaque nouveau paquet de données, l'algorithme optimise ses paramètres afin de se règler un peu mieux.

Ceci est toujours vrai en Machine Learning, quel que soit la librairie utilisée.



Pour cela, nous utilisons un noeud nommé **placeholder** qui ne réalise **aucun** calcul mais qui est chargé d'alimenter, d'emettre des données à l'exécution (*run time*).

Un exemple:






````
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
a = tf.placeholder(tf.float32)
````

On ne précise pas la taille du placeholder si on utilise :
```
shape=(None)
```



````
a = tf.placeholder(tf.float32 , None)
````

````
a = tf.placeholder(tf.float32 , None)
print(type(a))
print (a)

# prépare initialisation
init = tf.global_variables_initializer()

print("")
with tf.Session() as sess:    
  # lancer init 
  init.run() 

  data_1 = a.eval(feed_dict = {a : 1})
  print(data_1)

  print("")

  data_2 = a.eval(feed_dict = {a : [1. , 10.]})
  print(data_2)
  print(data_2.shape)  

  print("")

  data_3 = a.eval(feed_dict = {a : [[1.987 , 10] , [2.345 , 20]]})
  print(data_3)  
  print(data_3.shape)  
````

```feed_dict``` est un dictionnaire (structure de données de type *clé - valeur*). 

Le nom du placeholder (ici *a*) est la clé. 

Les données sont "émises" à l'exécution
